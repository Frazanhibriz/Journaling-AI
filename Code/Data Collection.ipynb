## Dataset 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/meisaputri21/Indonesian-Twitter-Emotion-Dataset/refs/heads/master/Twitter_Emotion_Dataset.csv')

df.head()
```

```python
kamus = pd.read_csv("https://raw.githubusercontent.com/meisaputri21/Indonesian-Twitter-Emotion-Dataset/refs/heads/master/kamus_singkatan.csv", sep=";", header=None, names=["singkatan", "hasil"])
kamus_dict = dict(zip(kamus['singkatan'], kamus['hasil']))
```

```python
def normalisasi_singkatan(teks):
    words = str(teks).split()
    return " ".join([kamus_dict[w.lower()] if w.lower() in kamus_dict else w for w in words])

df['tweet'] = df['tweet'].astype(str).apply(normalisasi_singkatan)
```

```python
from collections import Counter

label_counts = Counter(df['label'])
print("\nDistribusi Label Emosi:")
for label, count in label_counts.items():
    print(f"{label}: {count}")
```

```python
plt.bar(label_counts.keys(), label_counts.values())
plt.title("Distribusi Label Emosi")
plt.xlabel("Emosi")
plt.ylabel("Jumlah Data")
plt.show()
```

## Dataset 2

### Anger

```python
anger_data = pd.read_csv(
    "https://raw.githubusercontent.com/Ricco48/Emotion-Dataset-from-Indonesian-Public-Opinion/refs/heads/main/Emotion%20Dataset%20from%20Indonesian%20Public%20Opinion/AngerData.csv",
    sep="\t",
    on_bad_lines='skip'
)

print(anger_data.head())
print(anger_data.info())

anger_data = anger_data.rename(columns={"Tweet": "tweet", "Label": "label"})
anger_data['label'] = anger_data['label'].str.lower()


print(anger_data['label'].value_counts())
```

### Sad

## Concat

```python
df = df.rename(columns={"label": "label", "tweet": "tweet"})
df['label'] = df['label'].str.lower()

# anger
anger_data = anger_data.rename(columns={"Tweet": "tweet", "Label": "label"})
anger_data['label'] = anger_data['label'].str.lower()

# fear
fear_data = fear_data.rename(columns={"Tweet": "tweet", "Label": "label"})
fear_data['label'] = fear_data['label'].str.lower()

# joy
joy_data = joy_data.rename(columns={"Tweet": "tweet", "Label": "label"})
joy_data['label'] = joy_data['label'].str.lower()

# love
love_data = love_data.rename(columns={"Tweet": "tweet", "Label": "label"})
love_data['label'] = love_data['label'].str.lower()

# sad
sad_data = sad_data.rename(columns={"Tweet": "tweet", "Label": "label"})
sad_data['label'] = sad_data['label'].str.lower()

df['label'] = df['label'].replace("sadness", "sad")
df['label'] = df['label'].replace("happy", "joy")

df_all = pd.concat([df, anger_data, fear_data, joy_data, love_data, sad_data], ignore_index=True)
df_all = df_all.drop_duplicates().reset_index(drop=True)

print(df_all.head())
print()
print(df_all['label'].value_counts())
```

```python
label_counts_all = df_all['label'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(label_counts_all.index, label_counts_all.values)
plt.title("Distribusi Label Emosi (Combined Dataset)")
plt.xlabel("Emosi")
plt.ylabel("Jumlah Data")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
